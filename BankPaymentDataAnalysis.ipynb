```python
maxvalue=Bank[['no_atms_on_site','no_atms_off_site','no_pos_on_line','no_pos_off_line','no_credit_cards','no_credit_card_atm_txn',
               'no_credit_card_pos_txn','no_debit_cards','no_debit_card_atm_txn','no_debit_card_pos_txn']].mean()
print("maximum value is:")
print(maxvalue)

maximum=np.max(maxvalue)
print("\nThe Maximum value is::",maximum)

print("\nWe can predict that debit card is the most used methods by users")
```

```python
Bank_by_year=Bank[['year','no_atms_on_site', 'no_atms_off_site', 'no_pos_on_line',
       'no_pos_off_line', 'no_credit_cards', 'no_credit_card_atm_txn',
       'no_credit_card_pos_txn', 'no_debit_cards', 'no_debit_card_atm_txn',
       'no_debit_card_pos_txn']].groupby('year').aggregate(np.sum)
Bank_by_year.plot()
```
