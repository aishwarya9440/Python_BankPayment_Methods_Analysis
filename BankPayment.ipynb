# Visualization Of Bank Payment Dataset

### Pair Plot

```python
sb.pairplot(Bank)
```

### Scatter Plot

```python
sb.scatterplot(x='year',y='no_debit_card_atm_txn',data=Bank)
```

### Matplotlib Scatter plot for number of atms onsite against offsite.

```python
#create axis and figure
fig,ax=plt.subplots()

#scatter the no.of atm offsite against no.of atm onsite
ax.scatter(Bank['no_atms_on_site'],Bank['no_atms_off_site'])

#set a title and labels
ax.set_title('Bank Payment Data')
ax.set_xlabel('no_atms_on_site')
ax.set_ylabel('no_atms_off_site')
```

### Matplotlib Scatter plot for number of credit cards against debit cards.

```python
#create axis and figure
fig,ax=plt.subplots()

#scatter the no.of atm offsite against no.of atm onsite
ax.scatter(Bank['no_credit_cards'],Bank['no_debit_cards'])

#set a title and labels
ax.set_title('Bank Payment Data')
ax.set_xlabel('no_credit_cards')
ax.set_ylabel('no_debit_cards')
```

### Line chart

```python
#get column to plot
columns=['no_credit_cards','no_debit_cards']
#create x data
x_data=range(0,Bank.shape[0])

#create figure and axis
fig,ax=plt.subplots()

#plot each column
for column in columns:
    ax.plot(x_data,Bank[column])
    
#set title and legend
ax.set_title('Bank Payment Data')
```

### Histogram for credit cards reviews according to month.

```python
#craete a fig and axis
fig,ax=plt.subplots()

#plot Histogram
ax.hist(Bank['year'])

#set title and labels
ax.set_title('Credit Card Review')
ax.set_xlabel('month')
ax.set_ylabel('no_credit_cards')
```

### Creating multiple Histogram.

```python
Bank.plot.hist(subplots=True,layout=(5,5),figsize=(20,20),bins=20)
```

### Scatter Plot

```python
sb.relplot(x='year',y='no_credit_cards',hue='month',data=Bank)
```

### Bar Chart for number of debit cards per year.

```python
#create a figure and axis
fig,ax=plt.subplots()

#count the occurence of each class
data=Bank['year'].value_counts()

#get x and y data
year=data.index
no_debit_cards=data.values

#create bar chart
ax.bar(year,no_debit_cards)

#set title and labels
ax.set_title('Debit Card Review')
ax.set_xlabel('Year')
ax.set_ylabel('no_debit_cards')
```

### Bar Chart for number of atms off site per month..

```python
#create a figure and axis
fig,ax=plt.subplots()

#count the occurence of each class
data=Bank['month'].value_counts()

#get x and y data
month=data.index
no_atms_off_site=data.values

#create bar chart
ax.bar(month,no_atms_off_site)

#set title and labels
ax.set_title('ATM Review')
ax.set_xlabel('Month')
ax.set_ylabel('no_atms_Off_site')
```

### Bar chart(Year with most numbers of atm cards on site used).

```python

Bank.groupby("year").no_atms_on_site.mean().sort_values(ascending=False)[:5].plot.bar()
```
